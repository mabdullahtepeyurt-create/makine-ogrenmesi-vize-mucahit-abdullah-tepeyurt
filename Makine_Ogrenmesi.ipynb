```python
!pip install xgboost shap
#1.KÜTÜPHANELERİN İÇE AKTARILMASI

import numpy as np              #numpy kütüphanesinden NP parametresini çağırdım
import pandas as pd             #Veri manipülasyonu ve analizi için Pandas kütüphanesini (özellikle DataFrame yapıları için) 'pd' takma adıyla içe aktarır.
import matplotlib.pyplot as plt #Grafik ve çizim oluşturmak için Matplotlib kütüphanesinin Pyplot modülünü 'plt' takma adıyla içe aktarır.
import seaborn as sns           #Daha çekici ve bilgilendirici istatistiksel grafikler oluşturmak için Seaborn kütüphanesini içe aktarır (Matplotlib üzerine kurulu).
import copy                     #Nesneleri kopyalamak (özellikle derin kopyalama - deepcopy) için 'copy' modülünü içe aktarır. Bu, orijinal nesneyi etkilemeden model şablonlarını kullanmak için önemlidir.

from sklearn.datasets import load_breast_cancer       #Veri Seti Yükleme: Scikit-learn'ün hazır veri setlerinden Meme Kanseri (Breast Cancer) veri setini projeye dahil etmek için.
from sklearn.model_selection import train_test_split  #Veri Bölme: Veri setini eğitim ve test kümelerine rastgele ve kontrollü bir şekilde ayırmak için kullanılır.

from sklearn.preprocessing import StandardScaler                      #Ölçeklendirme (Scaling): Veriyi standart normal dağılıma dönüştürmek (μ=0,σ=1), yani Standardizasyon yapmak için kullanılır. Bu, uzaklığa dayalı algoritmalar için kritiktir."
from sklearn.decomposition import PCA                                 #Boyut Azaltma: Temel Bileşenler Analizi (Principal Component Analysis) algoritmasını içe aktarır. Varyansı koruyarak veri boyutunu düşürmek için kullanılır.
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis  #Boyut Azaltma: Doğrusal Diskriminant Analizi (LDA) algoritmasını içe aktarır. Sınıf ayrımını maksimize edecek şekilde veri boyutunu düşürmek için kullanılır (gözetimli bir yöntemdir).


from sklearn.linear_model import LogisticRegression #Doğrusal Model: İkili sınıflandırma problemleri için Lojistik Regresyon sınıflandırıcısını içe aktarır.
from sklearn.tree import DecisionTreeClassifier     #Ağaç Tabanlı Model: Karar verme kurallarına dayalı bir sınıflandırıcı olan Karar Ağacı modelini içe aktarır.
from sklearn.ensemble import RandomForestClassifier #Ensemble Model: Birden çok karar ağacını (topluluğunu) kullanarak daha sağlam ve genellenebilir tahminler yapan Rastgele Orman sınıflandırıcısını içe aktarır.
from xgboost import XGBClassifier                   #Ensemble Model: Yüksek performanslı ve ölçeklenebilir bir Gradyan Artırma (Gradient Boosting) algoritması olan XGBoost sınıflandırıcısını içe aktarır.
from sklearn.naive_bayes import GaussianNB          #Olasılıksal Model: Bayes teoremine dayanan, özellikle yüksek boyutlu verilerde iyi çalışan Gauss Naive Bayes sınıflandırıcısını içe aktarır.

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    roc_curve,
    auc
)

import shap  # XAI için

#Grafiklerin daha güzel görülebilmesi için yapılan uygulama
plt.rcParams['figure.figsize'] = (8, 6)
sns.set(style="whitegrid")

#2. VERİ SETİNİN YÜKLENMESİ (Breast Cancer Wisconsin)
data = load_breast_cancer()

#X → özellikler, y → hedef (sınıf etiketi)
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target, name="target")

print("Özellik matrisi boyutu (X):", X.shape)
print("Hedef vektörü boyutu (y):", y.shape)
print("\nSınıf isimleri:", data.target_names)

print("\nİlk 5 satır:")
display(X.head())

#3. VERİ KALİTE KONTROLÜ
#3.1 Eksik Değer Analizi
print("=== Eksik Değer Analizi ===")
missing_counts = X.isnull().sum()
print(missing_counts)

if missing_counts.sum() == 0:
    print("\nBu veri setinde eksik değer yok.")
else:
    print("\nEksik değer bulunan sütunlar yukarıda listelendi.")
    # Örnek: ortalama ile doldurma
    X = X.fillna(X.mean())
    print("Eksik değerler sütun ortalaması ile dolduruldu.")

#3.2 Aykırı Değer Analizi (Z-score yöntemi)
from scipy.stats import zscore

print("\n=== Aykırı Değer Analizi (Z-score) ===")
z_scores = np.abs(zscore(X))
outlier_indices = np.where(z_scores > 3)
print("Toplam aykırı değer sayısı:", len(outlier_indices[0]))

#Not: Bu projede aykırı değerleri silmiyoruz, sadece raporluyoruz.
#Çünkü bazı modeller; örneğin ağaç tabanlı, aykırılara daha dayanıklıdır.

#3.3 Veri Tipi ve Dağılım İncelemesi
print("\n=== Veri Tipleri ===")
print(X.dtypes)

numeric_cols = X.select_dtypes(include=[np.number]).columns
categorical_cols = X.select_dtypes(exclude=[np.number]).columns

print("\nSayısal değişken sayısı:", len(numeric_cols))
print("Kategorik değişken sayısı:", len(categorical_cols))

#4. KEŞİFSEL VERİ ANALİZİ (EDA)
print("=== Temel İstatistiksel Özellikler ===")
stats = X.describe(percentiles=[0.25, 0.5, 0.75]).T
display(stats[['mean', '50%', 'min', 'max']])

#4.1 Q1, Q3, IQR hesaplayalım
Q1 = X.quantile(0.25)
Q3 = X.quantile(0.75)
IQR = Q3 - Q1

eda_table = pd.DataFrame({
    "Mean": X.mean(),
    "Median": X.median(),
    "Min": X.min(),
    "Max": X.max(),
    "Std": X.std(),
    "Q1": Q1,
    "Q3": Q3
})

print("\n=== EDA Tablosu (Özet) ===")
display(eda_table.head())

#4.2 Korelasyon Matrisi ve Heatmap
corr_matrix = X.corr(method='pearson')

plt.figure()
sns.heatmap(corr_matrix, cmap="coolwarm", center=0)
plt.title("Korelasyon Matrisi Heatmap")
plt.tight_layout()
plt.show()

#En yüksek korelasyonlu 3 özellik çifti (Mutlak Değere Göre)
corr_pairs = corr_matrix.abs().where(
    np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)
).stack()

top3_corr = corr_pairs.sort_values(ascending=False).head(3)
print("\nEn yüksek korelasyona sahip 3 özellik çifti:")
print(top3_corr)

#4.3 Boxplot Analizi
plt.figure(figsize=(12, 6))
sns.boxplot(data=X)
plt.xticks(rotation=90)
plt.title("Tüm Özellikler İçin Boxplot")
plt.tight_layout()
plt.show()

for col in X.columns:
   plt.figure()
   sns.boxplot(x=X[col])
   plt.title(f"Boxplot - {col}")
   plt.show()

#5. VERİ ÖLÇEKLENDİRME (SCALING)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Ölçeklendirilmiş verinin şekli:", X_scaled.shape)

#6. VERİNİN EĞİTİM / VALIDATION / TEST OLARAK BÖLÜNMESİ
#Oranlar: %70 Train, %10 Val, %20 Test
#Önce: Train+Val ve Test ayır
X_temp, X_test, y_temp, y_test = train_test_split(
    X_scaled, y, test_size=0.20, random_state=42, stratify=y
)

#Sonra: Train ve Validation'ı ayır
#Train oranı: 0.7 / 0.8 = 0.875
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.125, random_state=42, stratify=y_temp
)

print("Train şekli:", X_train.shape)
print("Validation şekli:", X_val.shape)
print("Test şekli:", X_test.shape)

#7. PCA (Principal Component Analysis)
#7.1 PCA'yı önce tam boyutta Fit edelim (TÜM BİLEŞENLER)
pca_full = PCA()
pca_full.fit(X_train)

explained_var_ratio = pca_full.explained_variance_ratio_
mean_explained = explained_var_ratio.mean()

print("Explained variance oranları:\n", explained_var_ratio)
print("\nMean explained variance ratio:", mean_explained)

#Ortalama explained variance'tan büyük olan bileşen sayısını seç
n_components_pca = np.sum(explained_var_ratio > mean_explained)
print("\nSeçilen PCA bileşen sayısı:", n_components_pca)

#7.2 Seçilen bileşen sayısıyla PCA'yi yeniden Fit et
pca = PCA(n_components=n_components_pca)
X_train_pca = pca.fit_transform(X_train)
X_val_pca = pca.transform(X_val)
X_test_pca = pca.transform(X_test)

print("PCA-Train şekli:", X_train_pca.shape)

#7.3 Explained variance grafiği
plt.figure()
plt.plot(range(1, len(explained_var_ratio) + 1), explained_var_ratio, marker='o')
plt.axhline(mean_explained, linestyle='--', label='Mean explained variance')
plt.xlabel("Bileşen No")
plt.ylabel("Explained Variance Ratio")
plt.title("PCA Explained Variance Ratios")
plt.legend()
plt.tight_layout()
plt.show()

#7.4 İlk iki PCA bileşeni ile 2D scatter plot
plt.figure()
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_train, cmap="coolwarm", alpha=0.7)
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("PCA - İlk İki Bileşen ile Sınıf Ayrımı")
plt.tight_layout()
plt.show()

#8. LDA (Linear Discriminant Analysis)
lda = LinearDiscriminantAnalysis(n_components=1)

X_train_lda = lda.fit_transform(X_train, y_train)
X_val_lda = lda.transform(X_val)
X_test_lda = lda.transform(X_test)

print("LDA-Train şekli:", X_train_lda.shape)

#İlk iki LDA bileşeni ile scatter plot
plt.figure()
plt.scatter(X_train_lda[:, 0], y_train, c=y_train, cmap="coolwarm")
plt.xlabel("LDA 1")
plt.ylabel("Sınıf")
plt.title("LDA - Tek Bileşen ile Sınıf Ayrımı")
plt.show()

#9. MODELLERİN TANIMLANMASI
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=200, random_state=42),
    "XGBoost": XGBClassifier(
        n_estimators=300,
        learning_rate=0.05,
        max_depth=4,
        subsample=0.9,
        colsample_bytree=0.9,
        objective="binary:logistic",
        eval_metric="logloss",
        random_state=42,
        n_jobs=-1
    ),
    "Naive Bayes": GaussianNB()
}

#10. MODEL EĞİTİMİ VE VALIDATION ÜZERİNDE DEĞERLENDİRME
#(Ham veri, PCA verisi, LDA verisi → toplam 15 model)
def evaluate_model(name, model, X_tr, y_tr, X_v, y_v):
    """Bir modeli eğitip validation üzerinde metriklerini hesaplar."""
    #Modeli eğit
    model.fit(X_tr, y_tr)

    #Validation üzerinde tahmin yap
    y_pred = model.predict(X_v)

    #ROC-AUC için olasılık veya decision function
    if hasattr(model, "predict_proba"):
        y_proba = model.predict_proba(X_v)[:, 1]
    elif hasattr(model, "decision_function"):
        y_proba = model.decision_function(X_v)
    else:
        #Eğer yoksa, ROC-AUC hesaplayamıyoruz
        y_proba = None

    #Temel metrikler
    acc = accuracy_score(y_v, y_pred)
    prec = precision_score(y_v, y_pred)
    rec = recall_score(y_v, y_pred)
    f1 = f1_score(y_v, y_pred)

    #ROC-AUC varsa hesapla
    if y_proba is not None:
        roc = roc_auc_score(y_v, y_proba)
    else:
        roc = np.nan

    return {
        "model": name,
        "accuracy": acc,
        "precision": prec,
        "recall": rec,
        "f1": f1,
        "roc_auc": roc,
        "trained_model": model
    }

#Sonuçlar ve eğitilmiş modelleri saklamak için list/sözlük
results = []
trained_models = {}

#3 veri temsili: Ham veri (Raw), PCA, LDA
#NOT: X_train_lda, X_val_lda, X_test_lda LDA kısmında oluşturulmuş olmalı
representations = {
    "Raw": (X_train,     X_val,     X_test),
    "PCA": (X_train_pca, X_val_pca, X_test_pca),
    "LDA": (X_train_lda, X_val_lda, X_test_lda)
}

#Tüm temsiller ve tüm modeller üzerinde dön
for rep_name, (X_tr_rep, X_val_rep, X_te_rep) in representations.items():
    for model_name, model_template in models.items(): # Renamed 'model' to 'model_template'
        print(f"Eğitiliyor → Representation: {rep_name}, Model: {model_name}")

        #Her döngüde modelin yeni bir kopyasını oluştur
        model_instance = copy.deepcopy(model_template)

        #y_tr yerine y_train kullanıyorum. evaluate_model fonksiyonu X_tr'i ikinci argüman olarak alıyor.
        #Fonksiyon tanımı: def evaluate_model(name, model, X_tr, y_tr, X_v, y_v):
        res = evaluate_model(model_name, model_instance, X_tr_rep, y_train, X_val_rep, y_val) # Pass model_instance and corrected y
        res["representation"] = rep_name

        #Eğitimli modeli sakla (test ve SHAP vs. için)
        trained_models[(rep_name, model_name)] = res["trained_model"]

        results.append(res)

#Sonuçları DataFrame'e çevir
results_df = pd.DataFrame(results)
results_df = results_df[[
    "representation", "model", "accuracy", "precision", "recall", "f1", "roc_auc"
]]

print("\n=== VALIDATION SONUÇ TABLOSU ===")
display(results_df.sort_values(by="f1", ascending=False))

#11. EN İYİ MODELİN SEÇİLMESİ (VALIDATION SONUÇLARINA GÖRE)
#Önce F1 skoruna göre, sonra roc_auc'ya göre sıralayalım
best_row = results_df.sort_values(
    by=["f1", "roc_auc"], ascending=False
).iloc[0]

best_rep = best_row["representation"]
best_model_name = best_row["model"]

print("En iyi representation:", best_rep)
print("En iyi model:", best_model_name)
print("\nEn iyi model validation metrikleri:")
display(best_row)

#12. EN İYİ MODELİN TEST SETİ ÜZERİNDE DEĞERLENDİRİLMESİ
#0) Önce 10. bölümün çalıştığından emin ol:
#- results_df
#- trained_models
#- representations
#tanımlı olmalı.

print("results_df var mı? ->", 'results_df' in globals())
print("trained_models var mı? ->", 'trained_models' in globals())
print("representations var mı? ->", 'representations' in globals())

#Eğer yukarıda False görürsen, önce 10. bölümü çalıştırman gerekiyor.
#1) Validation sonuçlarından EN İYİ satırı bul (F1-score'a göre)
best_idx = results_df['f1'].idxmax()
best_row = results_df.loc[best_idx]

best_rep = best_row["representation"]   #Raw / PCA / LDA
best_model_name = best_row["model"]     #Örn: "LogReg", "SVM", "RF" vs.

print("\nSeçilen en iyi temsil (representation):", best_rep)
print("Seçilen en iyi model:", best_model_name)

#2) En iyi eğitimli modeli al
best_model = trained_models[(best_rep, best_model_name)]

#3) İlgili temsilin test verisini al
X_train_best, X_val_best, X_test_best = representations[best_rep]

#4) Test set üzerinde tahmin yap
y_test_pred = best_model.predict(X_test_best)

#Olasılık / decision function
if hasattr(best_model, "predict_proba"):
    y_test_proba = best_model.predict_proba(X_test_best)[:, 1]
elif hasattr(best_model, "decision_function"):
    y_test_proba = best_model.decision_function(X_test_best)
else:
    y_test_proba = None

#5) Test metrikleri
test_acc = accuracy_score(y_test, y_test_pred)
test_prec = precision_score(y_test, y_test_pred)
test_rec = recall_score(y_test, y_test_pred)
test_f1 = f1_score(y_test, y_test_pred)

#ROC-AUC hesaplanırken bazen test setinde tek sınıf kalabiliyor,
#bu durumda hata vermesin diye kontrol eklememiz gerekiyor:
if (y_test_proba is not None) and (len(np.unique(y_test)) > 1):
    test_roc = roc_auc_score(y_test, y_test_proba)
else:
    test_roc = np.nan

print("\n=== EN İYİ MODELİN TEST PERFORMANSI ===")
print("Accuracy :", test_acc)
print("Precision:", test_prec)
print("Recall   :", test_rec)
print("F1-score :", test_f1)
print("ROC-AUC  :", test_roc)
#12.1 Confusion Matrix (Test Seti)
cm = confusion_matrix(y_test, y_test_pred)

plt.figure()
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Tahmin Edilen")
plt.ylabel("Gerçek")
plt.title(f"Confusion Matrix - {best_model_name} ({best_rep})")
plt.tight_layout()
plt.show()
#12.2 ROC Eğrisi (Test Seti)
if y_test_proba is not None:
    fpr, tpr, thresholds = roc_curve(y_test, y_test_proba)
    roc_auc_val = auc(fpr, tpr)

    plt.figure()
    plt.plot(fpr, tpr, label=f"ROC curve (AUC = {roc_auc_val:.3f})")
    plt.plot([0, 1], [0, 1], linestyle="--")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"ROC Eğrisi - {best_model_name} ({best_rep})")
    plt.legend(loc="lower right")
    plt.tight_layout()
    plt.show()

    print("Not: Threshold değiştikçe FPR ve TPR değişir.")
    print("Daha düşük threshold → Daha fazla pozitif tahmin, yüksek recall, düşük precision.")
    print("Daha yüksek threshold → Daha az pozitif tahmin, yüksek precision, düşük recall.")
else:
    print("Bu model için ROC eğrisi hesaplanamadı (olasılık / decision_function yok).")

#13. XAI - SHAP ANALİZİ (En İyi Model İçin)
#SHAP için bir örnek alt küme seçelim (hesaplama hızlı olsun diye yaptım)
if best_rep == "Raw":
    X_shap = pd.DataFrame(X_test_best, columns=data.feature_names)
elif best_rep == "PCA":
    X_shap = pd.DataFrame(X_test_best, columns=[f"PCA_{i+1}" for i in range(X_test_best.shape[1])])
elif best_rep == "LDA":
    X_shap = pd.DataFrame(X_test_best, columns=[f"LDA_{i+1}" for i in range(X_test_best.shape[1])])

X_shap_sample = X_shap.sample(n=min(100, len(X_shap)), random_state=42)

#Tree tabanlı modeller için TreeExplainer daha uygundur (Random Forest, XGBoost vs.)
if isinstance(best_model, (RandomForestClassifier, XGBClassifier)):
    explainer = shap.TreeExplainer(best_model)
else:
    explainer = shap.Explainer(best_model, X_shap_sample)

shap_values = explainer(X_shap_sample)

#Eğer çıktı çok boyutluysa (binary class), class 1'i alalım
if isinstance(shap_values.values, np.ndarray) and shap_values.values.ndim == 3:
    shap_values_plot = shap_values[..., 1]
else:
    shap_values_plot = shap_values

shap.initjs()

print("=== SHAP Summary Plot (Dot) ===")
shap.summary_plot(shap_values_plot, X_shap_sample)

print("=== SHAP Summary Plot (Bar) ===")
shap.summary_plot(shap_values_plot, X_shap_sample, plot_type="bar")

#14. PCA ve LDA Temsilleri İçin EK SHAP KARŞILAŞTIRMASI
def shap_for_representation(rep_name, model_name):
    print(f"\n=== SHAP Analizi → Representation: {rep_name}, Model: {model_name} ===")

    model = trained_models[(rep_name, model_name)]
    X_tr, X_v, X_te = representations[rep_name]

    if rep_name == "Raw":
        cols = data.feature_names
    else:
        cols = [f"{rep_name}_{i+1}" for i in range(X_te.shape[1])]

    X_df = pd.DataFrame(X_te, columns=cols)
    X_sample = X_df.sample(n=min(100, len(X_df)), random_state=42)

    if isinstance(model, (RandomForestClassifier, XGBClassifier)):
        explainer = shap.TreeExplainer(model)
    else:
        explainer = shap.Explainer(model, X_sample)

    shap_vals = explainer(X_sample)

    if isinstance(shap_vals.values, np.ndarray) and shap_vals.values.ndim == 3:
        shap_vals_plot = shap_vals[..., 1]
    else:
        shap_vals_plot = shap_vals

    shap.summary_plot(shap_vals_plot, X_sample, show=True)
    shap.summary_plot(shap_vals_plot, X_sample, plot_type="bar", show=True)

#Örnek: Aynı en iyi model için PCA ve LDA üzerinde SHAP
#Eğer en iyi model Raw üzerinde ise, yine de aynı algoritmayı PCA ve LDA için kullanabiliriz.
algo_for_shap = best_model_name  #aynı algoritma

for rep in ["Raw", "PCA", "LDA"]:
    shap_for_representation(rep, algo_for_shap)



```
